```python
import torch # requires pytorch
import numpy as np
#import autograd
```

```python
device = torch.device('cpu')
r_vec = np.repeat([np.random.uniform(low=0.95,high=1.5,size=1000)],3,axis=0)
print(r_vec)
```

```python
x = torch.tensor(r_vec, requires_grad=True)#.cuda().detach().requires_grad_()
```

```python
rr = torch.sqrt(torch.sum(x.pow(2),axis=0))

@torch.enable_grad()
def saito77(r):
    return r.pow(2.5)+ r.pow(3.5)

@torch.enable_grad()
def dsaito77_dr(r):
    return 2.5*r.pow(1.5)+ 3.5*r.pow(2.5)
#y = 1.36e6 * rr.pow(-2.14) + 1.68e8 * rr.pow(-6.13)
```

```python
Ne=saito77(rr)
#Ne = 1.36e6 * rr**(-2.14) + 1.68e8 * rr**(-6.13)

#print(Ne)
```

```python
Ne.backward((Ne*0+1))
```
